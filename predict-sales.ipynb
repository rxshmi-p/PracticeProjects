```python
items = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/items.csv")
item_categories = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/item_categories.csv")
sales_train = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv")
test = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/test.csv")
shops = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/shops.csv")
sample_submission = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/sample_submission.csv")
```

```python
print(items.columns)
print(item_categories.columns)
print(shops.columns)
print(sales_train.columns)
print(test.columns)
```

```python
features = ['ID', 'shop_id', 'item_id', 'item_category_id', 'item_category_name', 'item_cnt_day', 'item_price', 'date', 'date_block_num', 'item_name', 'shop_name']

info = pd.DataFrame(columns = features)

info.ID = test.ID
info.shop_id = test.shop_id
info.item_id = test.item_id
# for i in features :
#    if i in items.columns
```

```python
df_items = []
df_names = []
df_item_category_name = []

i = 0

for i in info.item_id :
    if i in items.item_id :
        df_items.append(items.item_category_id[i])
        df_names.append(items.item_name[i])
        i += 1

info.item_category_id = df_items
info.item_name = df_names

for i in info.item_category_id : 
    if i in item_categories.item_category_id : 
        df_item_category_name.append(item_categories.item_category_name[i])
        i += 1

info.item_category_name = df_item_category_name
```

```python
df_item_cnt = []
i = 0

for i in info.item_id : 
    if i in sales_train.item_id : 
        df_item_cnt.append(sales_train.item_cnt_day[i])
        i += 1

info.item_cnt_day = df_item_cnt
sales_train.iloc[5233]
```

```python
df_price = []
df_date = []
df_date_block = []

i = 0

for i in info.item_id :
    if i in sales_train.item_id :
        df_date.append(sales_train.date[i])
        df_date_block.append(sales_train.date_block_num[i])
        df_price.append(sales_train.item_price[i])
        i += 1

info.item_price = df_price
info.date = df_date
info.date_block_num = df_date_block

sales_train.iloc[5037]
```

```python
df_shop_name = []

for i in info.shop_id : 
    if i in shops.shop_id : 
        df_shop_name.append(shops.shop_name[i])
        i += 1

info.shop_name = df_shop_name
```
